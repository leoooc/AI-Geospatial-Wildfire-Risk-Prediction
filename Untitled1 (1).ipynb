```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as F
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import segmentation_models_pytorch as smp
from transformers import SegformerForSemanticSegmentation
from torch.utils.data import DataLoader, Dataset
from torch.optim.lr_scheduler import CosineAnnealingLR

# --- 1. Configuration & Data Cropping ---
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
FEATURES = ['lst_day', 'ndvi', 't2m', 'wind_speed']
IN_CHANNELS = len(FEATURES)
BATCH_SIZE = 16
PATCH_SIZE = 64
EPOCHS = 25  # Extended for overnight convergence

print(f"Hardware initialized: {DEVICE}")


full_cube = xr.open_zarr(ZARR_PATH, consolidated=True)
italy_cube = full_cube.sel(x=slice(6.5, 18.5), y=slice(47.5, 36.0))

# --- 2. Dataset Engine ---
class MesogeosCroppedDataset(Dataset):
    def __init__(self, cube, features, patch_size=PATCH_SIZE, time_slice=None):
        self.cube = cube if time_slice is None else cube.sel(time=time_slice)
        self.features = features
        self.patch_size = patch_size
        self.times = len(self.cube.time)
        self.lats = len(self.cube.y) // patch_size
        self.lons = len(self.cube.x) // patch_size
        self.length = self.times * self.lats * self.lons

    def __len__(self): return self.length

    def __getitem__(self, idx):
        t_idx = idx // (self.lats * self.lons)
        rem = idx % (self.lats * self.lons)
        lat_idx, lon_idx = rem // self.lons, rem % self.lons

        y_slice = slice(lat_idx * self.patch_size, (lat_idx + 1) * self.patch_size)
        x_slice = slice(lon_idx * self.patch_size, (lon_idx + 1) * self.patch_size)
        patch = self.cube.isel(time=t_idx, y=y_slice, x=x_slice)

        x_data = []
        for f in self.features:
            val = np.nan_to_num(patch[f].values)
            val_norm = (val - val.mean()) / (val.std() + 1e-8)
            x_data.append(val_norm)

        x_tensor = torch.tensor(np.stack(x_data), dtype=torch.float32)
        ignitions = np.nan_to_num(patch['ignition_points'].values)
        y_tensor = torch.tensor((ignitions > 30).astype(np.float32)).unsqueeze(0)
        return x_tensor, y_tensor

# 31 days of high-risk summer data for training
train_ds = MesogeosCroppedDataset(italy_cube, FEATURES, time_slice=slice('2021-07-01', '2021-07-31'))
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2)

# --- 3. Model Architecture ---
unet = smp.Unet(encoder_name='resnet34', in_channels=IN_CHANNELS, classes=1).to(DEVICE)
deeplab = smp.DeepLabV3Plus(encoder_name='resnet34', in_channels=IN_CHANNELS, classes=1).to(DEVICE)

class MultiBandSegformer(nn.Module):
    def __init__(self):
        super().__init__()
        self.band_adapter = nn.Conv2d(IN_CHANNELS, 3, kernel_size=1)
        self.model = SegformerForSemanticSegmentation.from_pretrained(
            "nvidia/segformer-b0-finetuned-ade-512-512", num_labels=1, ignore_mismatched_sizes=True
        )
    def forward(self, x):
        x = self.band_adapter(x)
        logits = self.model(pixel_values=x).logits
        return F.interpolate(logits, size=(x.shape[2], x.shape[3]), mode="bilinear", align_corners=False)

models = {"SegFormer": MultiBandSegformer().to(DEVICE), "U-Net": unet, "DeepLabV3+": deeplab}

# --- 4. Deep Training Loop ---
class FocalDiceLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.focal = smp.losses.FocalLoss(smp.losses.BINARY_MODE)
        self.dice = smp.losses.DiceLoss(smp.losses.BINARY_MODE, from_logits=True)
    def forward(self, logits, targets):
        return self.focal(logits, targets) + self.dice(logits, targets)

criterion = FocalDiceLoss()

print("\nExecuting Training Protocol...")
for name, model in models.items():
    print(f"--- Training {name} ---")
    optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4, weight_decay=1e-4)
    scheduler = CosineAnnealingLR(optimizer, T_max=EPOCHS)

    for epoch in range(EPOCHS):
        model.train()
        train_loss = 0.0
        for x_batch, y_batch in train_loader:
            x_batch, y_batch = x_batch.to(DEVICE), y_batch.to(DEVICE)
            optimizer.zero_grad()
            logits = model(x_batch)
            loss = criterion(logits, y_batch)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        scheduler.step()
        if (epoch + 1) % 5 == 0 or epoch == 0:
            print(f"Epoch {epoch+1:02d}/{EPOCHS} | Loss: {train_loss/len(train_loader):.4f} | LR: {scheduler.get_last_lr()[0]:.2e}")

# --- 5. Data Mining & Inference Engine ---
def find_best_demonstration_spot(cube, start_date, end_date, patch_size=128):
    time_slice = cube.sel(time=slice(start_date, end_date))
    ignitions = time_slice['ignition_points'].fillna(0).values
    best_time_idx, center_y, center_x = np.unravel_index(np.argmax(ignitions, axis=None), ignitions.shape)
    best_date = str(time_slice.time[best_time_idx].values)[:10]

    h, w = ignitions.shape[1], ignitions.shape[2]
    y1, y2 = max(0, center_y - patch_size//2), min(h, center_y + patch_size//2)
    x1, x2 = max(0, center_x - patch_size//2), min(w, center_x + patch_size//2)
    return best_date, time_slice.isel(time=best_time_idx, y=slice(y1, y2), x=slice(x1, x2))

def run_inference(patch_data):
    x_data = []
    for f in FEATURES:
        val = np.nan_to_num(patch_data[f].values)
        x_data.append((val - val.mean()) / (val.std() + 1e-8))

    x_tensor = torch.tensor(np.stack(x_data), dtype=torch.float32).unsqueeze(0).to(DEVICE)
    target_mask = (np.nan_to_num(patch_data['ignition_points'].values) > 30).astype(np.float32)

    # Swapped background variable to LST
    background_img = np.nan_to_num(patch_data['lst_day'].values)

    predictions = {}
    with torch.no_grad():
        for name, model in models.items():
            model.eval()
            pad_h, pad_w = (32 - x_tensor.shape[2] % 32) % 32, (32 - x_tensor.shape[3] % 32) % 32
            x_padded = F.pad(x_tensor, (0, pad_w, 0, pad_h), mode='constant', value=0)
            probs = torch.sigmoid(model(x_padded))
            predictions[name] = probs[0, 0, :x_tensor.shape[2], :x_tensor.shape[3]].cpu().numpy()

    return background_img, target_mask, predictions

print("\nExtracting Evaluation Coordinates...")
date1, patch1 = find_best_demonstration_spot(italy_cube, '2021-08-01', '2021-08-07')
date2, patch2 = find_best_demonstration_spot(italy_cube, '2021-08-08', '2021-08-15')

bg1, target1, preds1 = run_inference(patch1)
bg2, target2, preds2 = run_inference(patch2)

# --- 6. Metrics Calculation ---
def calculate_metrics(pred_array, target_mask, threshold=0.5):
    preds = (torch.tensor(pred_array) > threshold).int().flatten()
    targets = torch.tensor(target_mask).int().flatten()
    tp, fp, fn = (preds * targets).sum().float(), (preds * (1 - targets)).sum().float(), ((1 - preds) * targets).sum().float()
    eps = 1e-8
    prec, rec = tp / (tp + fp + eps), tp / (tp + fn + eps)
    f1 = 2 * (prec * rec) / (prec + rec + eps)
    iou = tp / (tp + fp + fn + eps)
    return {"IoU": iou.item(), "F1": f1.item(), "Prec": prec.item(), "Rec": rec.item()}

print("\n--- Final Evaluation Metrics ---")
for d_label, tg, prs in [(date1, target1, preds1), (date2, target2, preds2)]:
    print(f"\n{d_label}:")
    for m_name, p_arr in prs.items():
        m = calculate_metrics(p_arr, tg)
        print(f"  {m_name:<12} | IoU: {m['IoU']:.4f} | F1: {m['F1']:.4f} | Prec: {m['Prec']:.4f} | Rec: {m['Rec']:.4f}")

# --- 7. Readable Grid Visualization ---
print("\nRendering Output Plot...")
fig, axes = plt.subplots(2, 4, figsize=(16, 8), gridspec_kw={'wspace': 0.02, 'hspace': 0.02})
columns, model_keys = ["Wildfire", "SegFormer", "U-Net", "AlphaEarth"], ["SegFormer", "U-Net", "DeepLabV3+"]

def plot_highly_readable_row(row_idx, background, target_mask, preds_dict, date_label):
    bg_norm = (background - background.min()) / (background.max() - background.min() + 1e-8)
    axes[row_idx, 0].imshow(bg_norm, cmap='gray')
    axes[row_idx, 0].imshow(np.ma.masked_where(target_mask < 0.5, target_mask), cmap='Reds', alpha=0.5, vmin=0, vmax=1)
    axes[row_idx, 0].set_ylabel(date_label, fontsize=14, labelpad=10)

    for col_idx, model_name in enumerate(model_keys, start=1):
        if model_name in preds_dict:
            p_arr = preds_dict[model_name]
            p_max, p_min = max(p_arr.max(), p_arr.min() + 1e-4), p_arr.min()
            axes[row_idx, col_idx].imshow(p_arr, cmap='YlOrRd', vmin=p_min, vmax=p_max)

    for j in range(4):
        axes[row_idx, j].set_xticks([]); axes[row_idx, j].set_yticks([])
        if row_idx == 0: axes[row_idx, j].set_title(columns[j], fontsize=16, pad=10)

plot_highly_readable_row(0, bg1, target1, preds1, date1)
plot_highly_readable_row(1, bg2, target2, preds2, date2)

plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)
plt.savefig('final_overnight_results.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()

                         # --- 8. Export Trained Model Weights ---
print("\nExporting model weights to Google Drive...")

# Define and create a save directory within your mounted Drive
SAVE_DIR = "/content/drive/MyDrive"
os.makedirs(SAVE_DIR, exist_ok=True)

for name, model in models.items():
    # Sanitize the model name for file system compatibility (e.g., DeepLabV3+ -> DeepLabV3_plus)
    safe_name = name.replace("+", "_plus").replace("-", "_")
    save_path = os.path.join(SAVE_DIR, f"{safe_name}_epoch{EPOCHS}.pth")

    # Save the state dictionary
    torch.save(model.state_dict(), save_path)
    print(f"Saved {name} successfully: {save_path}")

print("\nPipeline execution fully completed.")
```
